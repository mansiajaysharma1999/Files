# Scrape the details most watched tv series of all time from imdb.com.

Url = https://www.imdb.com/list/ls095964455/

We have to find the following details:
    
A) Name

B) Year span

C) Genre

D) Run time

E) Ratings

F) Votes

```python
#scrap the web page
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
#scrap the names of watched T.V serial
name=[]
n=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/a')
for i in n:
    name.append(i.text)
    
#scrap the year span of serial
year=[]
y=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/span[2]')
for i in y:
    year.append(i.text)

#scrap the genre of serials    
genre=[]
g=driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in g:
    genre.append(i.text)

#scrap the details of run_time    
run_time=[]
r=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in r:
    run_time.append(i.text) 

#scrap votes    
votes=[]
v=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]')
for i in v:
    votes.append(i.text)
    
#scrap ratings    
rating=[]
r=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div/span[2]')
for i in r:
    rating.append(i.text)    
```

```python
df=pd.DataFrame({'NAME':name,'YEAR SPAN':year,'GENRE':genre,'RUNTIME':run_time,'RATINGS':rating,'VOTES':votes})
df
```

```python
# fetching out web page
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://github.com/')
```

```python
# import necessary exceptions
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
```

```python
# click on open source
source=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
source.click()
```

```python
delay=5
```

```python
# scrap repositort title
title=[]
delay=5
try:
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,'//h2[@class="h3 lh-condensed"]')))
    t=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]')
    for i in t:
        title.append(i.text)
except TimeOutException as e:
    print("exception raised",e)
```

```python
title
```

```python
# scrap repository discription
disc=[]
try:
    d=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in d:
        disc.append(i.text)
except NoSuchElementException as e:
    disc.append("-")
```

```python
disc
```

```python
# scrap url
url=[]
u=driver.find_elements(By.XPATH,'//article[@class="Box-row"]/h2/a')
for i in u:
    url.append(i.get_attribute("href"))
```

```python
url
```

```python
# scrap contributors count:
cont=[]
for i in url:
    driver.get(i)
    time.sleep=2
    try:
        cont_tags = driver.find_element(By.XPATH,'//span[contains(@class="Counter ml-1")]')
        Cont.append(cont_tags.replace('//span[contains(@class="Counter ml-1")')
    except NoSuchElementException as e:
           cont.append('-')
                                      
             

```

```python
# scrap language that used in repository:
lang=[]
for i in url:
    driver.get(i)
    try:
        l=driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        lang.append(l.text)
    except NoSuchElementException as e:
        lang.append("--")
```

# Details of Datasetsfrom UCI machine learning repositories.

You have to find the following details:
A) Dataset name

D) Attribute type
E) No of instances
F) No of attribute
G) Year

```python
#scrap the web page
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://archive.ics.uci.edu/')
```

```python
# click on dataset
view=driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
view.click()
```

```python
# click on expand all
delay=5
try:
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,'//span[@class="swap-on text-primary-content"]')))
    expand_all=driver.find_element(By.XPATH,'//span[@class="swap-on text-primary-content"]')
    expand_all.click()
except TimeoutException as e: 
       print('raised error',e)
```

```python
# scrap the name of dataset
dataset=[]
d=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
for i in d:
    dataset.append(i.text)
    next_b=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]/svg')
```

```python
dataset
```

```python
# scrap types of attribute
attribute_t=[]
a=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
for i in a:
    attribute_t.append(i.text)
    next_b=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]/svg')
```

```python
# scrap year
year=[]
y=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')
for i in y:
    year.append(i.text)
    next_b=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]/svg')
        
```
